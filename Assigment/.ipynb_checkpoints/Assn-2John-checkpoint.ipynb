````python
import re

url = "https://www.dotabuff.com/esports/matches"
user_agent = {'User-agent': 'Mozilla/5.0'}

all_match_ids = []

def get_match_ids(sp):
    
    for i in sp.find_all('td'):
        for j in i.find_all('a'):
            a_link = str(j)
            m = re.search(r'[^>/]\d{9}(?=<)',a_link)
            if m != None:
                all_match_ids.append(int(m.group()))
                
    return all_match_ids

for i in range(1,51):
    if i == 1:
        source_code = requests.get(url, headers=user_agent).text
        soup = BeautifulSoup(source_code,'lxml')
        get_match_ids(soup)
    else:
        source_code = requests.get(url+"?page="+str(i), headers=user_agent).text
        soup = BeautifulSoup(source_code,'lxml')
        get_match_ids(soup)
        

            
# ?page=2
````

### 3.1 Find average win rate for each faction

Your answer should be numerical values

````python
# Fill code here
matches_read[0]['radiant_win']
````

````python
radiant_victory = 0
dire_victory = 0
for i,win_loss in enumerate(matches_read):
    try:
        if win_loss['radiant_win']:
            radiant_victory += 1
        else:
            dire_victory += 1
    except:
        pass
````

````python
total_games = float(radiant_victory + dire_victory)
radiant_win_rate = radiant_victory / total_games
dire_win_rate = dire_victory / total_games
````

````python
print "radiant win rate: ", radiant_win_rate
print "dire win rate: ", dire_win_rate
````

### 3.2 Game duration

Plot a histogram to show distribution of game duration. Also mark on the plot: the **mean** and **median**.

````python
# Fill code here
# duration_df = matches_read
all_duration = []
for duration in matches_read:
    all_duration.append(duration['duration'])
````

````python
all_duration = np.array(all_duration)/60.
duration_df = pd.DataFrame(all_duration)
duration_df.columns = ['duration']
duration_df['duration'] = duration_df['duration']
````

````python
to_plot = duration_df.plot.hist(bins=60,figsize=(11,6))
ymin = 0
ymax = 100
to_plot.vlines(x=duration_df.mean(), ymin=ymin, ymax=ymax-1, color='g', linestyles='dotted')
to_plot.vlines(x=duration_df.median(), ymin=ymin, ymax=ymax-1, color='k', linestyles='dashed')
to_plot.legend(['duration','mean: 41.91','median: 41.78'])
````

### 3.4 Hero popularity and win rate

Make a stacked horizontal barchart to display the number of games played by each hero and how many of them are victories. You must sort the bar so that the most popular hero appears on top.

````python
# Fill code here

def frequency(ls):
        dic = {}
        for hero in ls:
            if hero in dic.keys():
                dic[hero] += 1
            else:       
                key = hero
                value = 1
                dic[key] = value
        return dic
````

````python
# player 0 - 4 Radiant

picked = []
win = []
lose = []
for ele in matches_read:
    for j in ele['players']:
        try:
            hero = j['hero_name']
            rad_win = ele['radiant_win']
            picked.append(hero)
            if rad_win and j['player_slot'] < 5:
                win.append(hero)
            if (not rad_win) and j['player_slot'] > 100:
                lose.append(hero)
        except:
            pass
        
````

````python
print "win", len(win)
print "lose", len(lose)
print len(picked)
````

````python
all_win = frequency(win)
all_lose = frequency(lose)
all_picked = frequency(picked)
````

````python
all_win_df = pd.DataFrame(pd.Series(all_win))
all_picked_df = pd.DataFrame(pd.Series(all_picked))
````

````python
play_win_df = pd.concat([all_picked_df,all_win_df], ignore_index=True, axis=1)
play_win_df.columns = ['picked', 'win']
play_win_df.index.name = 'hero'
````

````python
play_win_df.sort_values(by=['picked'],ascending=False)[:30].plot(kind="barh",stacked=True, figsize=(14,8))
# .plot(kind='barh',stacked=True)
````

### 3.5  Kills/Death/Assist

Make a grouped horizonal barchart to display: total kills, total deaths, total assists for each hero on the same axis.

````python
# Fill code here

hero_kda = []
for ele in matches_read:
    for j in ele['players']:
        try:
            hero = j['hero_name']
            k = j['kills']
            d = j['deaths']
            a = j['assists']
            hero_kda.append({'hero':hero, 'kills':k, 'deaths':d, 'assists':a})
        except:
            pass
````

````python
hero_kda_df = pd.DataFrame(hero_kda)
````

````python
hero_kda_df = hero_kda_df.groupby('hero').count()
````

````python
hero_kda_df.sort_values(by=['kills'], ascending=False)[:30].plot(kind="barh",stacked=True,figsize=(14,8))
````

### 3.6 Richest Heros

Find out the top 10 heros with highest Gold per Minute.

````python
# Fill code here
hero_gold = []
for ele in matches_read:
    for j in ele['players']:
        try:
            hero = j['hero_name']
            gold = j['gold_per_min']
            hero_gold.append({'hero':hero, 'gold_per_min':gold})
        except:
            pass
````

````python
hero_gold_df = pd.DataFrame(hero_gold)
````

````python
hero_freq = frequency(list(hero_gold_df['hero']))
hero_freq = pd.DataFrame(pd.Series(hero_freq))
````

````python
gold_sum = hero_gold_df.groupby('hero').sum()
hero_gold_df = pd.concat([gold_sum,hero_freq],ignore_index=False, axis=1)
````

````python
hero_gold_df.columns = ['gold_per_min', 'freq']
hero_gold_df = hero_gold_df['gold_per_min'] / hero_gold_df['freq']
# hero_gold_df.columns = ['gold_per_min']
````

````python
hero_gold_df.sort_values()[-10:].plot(kind='barh')
````

### 3.6 Popular Items (overall)

Find out the top 10 items that most frequently used overall.

````python
# Fill code here
hero_item = []
for ele in matches_read:
    for j in ele['players']:
        try:
            for i in range(0,6):
                 hero_item.append({'item':j['item_'+ str(i) +'_name']})
        except:
            pass
````

````python
hero_item_df = pd.DataFrame(hero_item)
````

````python
hero_item_df = pd.DataFrame(hero_item_df.groupby('item').size())
hero_item_df.columns = ['freq']
````

````python
hero_item_df.sort_values(by=['freq'],ascending=False)[:10].plot(kind='barh')
````

### 3.7 Popular Items (among winners)

Find out the top 10 items that most frequently used among the winning team. 

````python
# Fill code here
vhero_item = []
for ele in matches_read:
    for j in ele['players']:
        try:
            rad_win = ele['radiant_win']
            if rad_win:
                for i in range(0,6):
                    vhero_item.append({'item':j['item_'+ str(i) +'_name']})
        except:
            pass
````

````python
vhero_item_df = pd.DataFrame(vhero_item)
vhero_item_df = pd.DataFrame(vhero_item_df.groupby('item').size())
vhero_item_df.columns = ['freq']
vhero_item_df.sort_values(by=['freq'],ascending=False)[:10].plot(kind='barh')
````
